```python
import numpy as np
import pandas as pd
import seaborn as sns
```

```python
df=pd.read_csv("https://github.com/campusx-official/100-days-of-machine-learning/raw/refs/heads/main/day44-outlier-detection-using-percentiles/weight-height.csv")
df.head()
```

```python
sns.kdeplot(df["Height"])
```

```python
sns.boxplot(df["Height"])
```

```python
df["Height"].describe()
```

```python
upper_limit=df["Height"].quantile(0.99)
upper_limit
```

```python
lower_limit=df["Height"].quantile(0.01)
lower_limit
```

```python
# outliers
df[(df["Height"]>upper_limit) | (df["Height"]<lower_limit)]
```

```python
df.shape
```

```python
# trimming
new_df=df[(df["Height"]<upper_limit)&(df["Height"]>lower_limit)]
```

```python
new_df.shape
```

```python
sns.boxplot(new_df["Height"])
```
